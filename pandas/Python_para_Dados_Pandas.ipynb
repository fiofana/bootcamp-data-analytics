**Extraindo estatísticas**

```python
print(df_cachorros['Altura (cm)'].min())
print(df_cachorros['Altura (cm)'].max())
print(df_cachorros['Nome'].max())
```

```python
def pct50(coluna):
    return coluna.quantile (0.50)

df_cachorros['Altura (cm)'].agg(pct50)
```

```python
df_cachorros[['Altura (cm)', 'Peso (kg)']].agg(pct50)
```

```python
def max_min(coluna):
    return coluna.max(), coluna.min()

df_cachorros [['Altura (cm)', 'Peso (kg)']].agg(max_min)
```

**Estatísticas cumulativas**

O primeiro número retornado, ou o número no índice zero, é o peso do primeiro cachorro. O próximo número é a soma do peso do primeiro e segundo cachorros. O terceiro número é a soma do peso do primeiro, segundo e terceiro cachorros, e assim por diante. O último número é a soma de todos os pesos dos cachorros.

```python
df_cachorros['Peso (kg)'].cumsum()
```

```python
df_cachorros['Peso (kg)'].cummin()
```

```python
df_cachorros['Peso (kg)'].cummax()
```

# Exemplo - Carros

```python
# Upload do CSV carros.csv
df_carros = pd.read_csv('carros.csv')

# Print do head com 10 linhas
df_carros.head(10)
```

```python
# Print das colunas e index do dataframe
print(df_carros.columns)
print(df_carros.index)
```

```python
# Sort decrescente no dataframe pelo estado com maior quantidade de carros
df_carros.sort_values('Quant Carros', ascending=False).head()
```

```python
# Proporção de carro para a população de cada estado
df_carros['Carros por habitante'] = df_carros['Quant Carros'] / df_carros['População do Estado']
df_carros.head()
```

```python
# Dataframe com Estado e proporção de carro por habitante
df_carros_habit = df_carros[['Estado', 'Carros por habitante']]
df_carros_habit.head()
```

```python
# Média de carros por região
df_carros.groupby('Região')['Quant Carros'].mean()
```

# Exemplo - Concurso

```python
df_concurso = pd.read_csv('concurso.csv')
df_concurso.head()
```

```python
# Porcentagem de pessoas com deficiência
pcds = df_concurso[df_concurso['Deficiência'] == 'Sim']['Número de Inscrição'].count()
pct_def = pcds / df_concurso['Número de Inscrição'].count() * 100
print(f'{pct_def:.2f}%')
```

```python
# resolução do video (???????????)
df_concurso['Deficiência'].value_counts(normalize=True) * 100
```

```python
# Pessoas do Piauí
df_concurso[df_concurso['Estado'] == 'PI']['Número de Inscrição'].count()
```

```python
# Distribuição de escolaridade por estado
df_concurso[['Estado', 'Escolaridade', 'Número de Inscrição']].groupby(['Estado', 'Escolaridade']).count()
```

```python
# Pessoas do ES que são do sexo feminino
capixabas = df_concurso[df_concurso['Estado'] == 'ES']
capixabas['Sexo'].value_counts()
```

```python
# Dados dos inscritos com as colunas: nº inscrição, nome e idade
df_concurso[['Número de Inscrição', 'Nome', 'Data de Nascimento']]
```
