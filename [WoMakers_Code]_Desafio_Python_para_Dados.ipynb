Você é uma pesquisadora que está tentando entender melhor qual o impacto
do estilo de vida de uma pessoa na sua qualidade de sono, por isso fez a
coleta dos dados de sobre 373 pessoas, onde foram recolhidas 12
características para cada uma delas. Por competência a sua pesquisa foi bem
controlada e você não tem dados faltosos na sua base. Chegou o momento
de você fazer sua análise e responder algumas perguntas.

```python
import pandas as pd
import numpy as np
```

```python
saude_sono = pd.read_csv('saude_do_sono_estilo_vida.csv')
saude_sono.head()
```

1. Ao visualizar a base você percebeu que seria melhor alterar o nome de
algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da
coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para
‘Profissão’, a coluna ‘Categoria BMI’ está em parte em inglês, substitua
para ‘Categoria IMC’.

```python
saude_sono.rename(columns={'ID': 'Identificador', 'Ocupação': 'Profissão', 'Categoria BMI': 'Categoria IMC', 'Pressão sanguíneaaaa': 'Pressão sanguínea'}, inplace=True)

saude_sono.head()
```

2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mod]

```python
media_sono = saude_sono.groupby('Profissão')['Duração do sono'].mean()
mediana_sono = saude_sono.groupby('Profissão')['Duração do sono'].median()
moda_sono = saude_sono.groupby('Profissão')['Duração do sono'].apply(pd.Series.mode)
```

```python
media_sono
```

```python
mediana_sono
```

```python
moda_sono
```

3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?

```python
eng_software = saude_sono[saude_sono['Profissão'] == 'Eng. de Software']
porcentagem_obesos = (eng_software['Categoria IMC'] == 'Obesidade').mean() * 100

print(f'A porcentagem de obesos em pessoas que atuam com engenharia de software é de {porcentagem_obesos}%.')
```

4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)

```python
media_sono_adv = saude_sono[saude_sono['Profissão'] == 'Advogado(a)']['Duração do sono'].mean()
media_sono_repr = saude_sono[saude_sono['Profissão'] == 'Representante de Vendas']['Duração do sono'].mean()

media_sono_adv > media_sono_repr
```

```python
print(f'Advogar faz você dormir menos.')
```

5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)

```python
profissoes_enf_med = ['Enfermeiro(a)', 'Médico(a)']
media_sono_enf_med = saude_sono[saude_sono['Profissão'].isin(profissoes_enf_med)].groupby('Profissão')['Duração do sono'].mean()
media_sono_enf_med
```

```python
print('Médicos têm menos horas de sono.')
```

7. Descubra qual a profissão menos frequente no conjunto. (Use
value_counts)

```python
profissao_menos_frequente = saude_sono['Profissão'].value_counts().idxmin()
print('A profissão menos frequente é', profissao_menos_frequente)
```

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
# Divisão da pressão sanguínea em dois valores e conversão para numérico
saude_sono[['Pressão sistólica', 'Pressão diastólica']] = saude_sono['Pressão sanguínea'].str.split('/', expand=True).astype(int)

media_pressao_genero = saude_sono.groupby('Gênero')[['Pressão sistólica', 'Pressão diastólica']].mean()
media_pressao_genero
```

9. É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?


```python
moda_duracao_sono = saude_sono['Duração do sono'].mode()[0]
predominante_8_horas = moda_duracao_sono == 8.0
predominante_8_horas
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
passos_acima_70 = saude_sono[saude_sono['Frequência cardíaca'] > 70]['Passos diários'].mean()
passos_abaixo_ou_igual_70 = saude_sono[saude_sono['Frequência cardíaca'] <= 70]['Passos diários'].mean()

passos_acima_70 > passos_abaixo_ou_igual_70
```
